```python
from scipy import misc
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Carregando a Imagem a ser tratada

```python
img = misc.imread("baboon.png")
```

### Dimens√µes da Imagem Carregada

```python
print (img.shape)
```

```python
plt.imshow(img, cmap="gray")
plt.show()
```

```python
print("largura:",img.shape[1])
print("altura:",img.shape[0])
print("intensidade m√≠nima:",img.min())
print("intensidade m√°xima:",img.max())
print("intensidade m√©dia: %0.2f" % img.mean())
```

```python
histogram_example = plt.hist(img)
plt.xlabel("N√≠veis de cinza")
plt.ylabel("Frequ√™ncia")
plt.show()
```

### Gerando o negativo

```python
img_neg = 255 - img
```

```python
plt.imshow(img_neg, cmap="gray")
plt.show()
```

### Aplicando a transforma√ß√£o sobre a imagem

```python
img_tr = ((60/255) * img) + 120
```

```python
plt.imshow(img_tr, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
print("largura:",img_tr.shape[1])
print("altura:",img_tr.shape[0])
print("intensidade m√≠nima: %0.f" % img_tr.min())
print("intensidade m√°xima: %0.f" % img_tr.max())
print("intensidade m√©dia: %0.f" % img_tr.mean())
```
