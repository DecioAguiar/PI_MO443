```python
img = misc.imread("baboon.png")
```

```python
print (img.shape, img.dtype)
```

```python
plt.imshow(img, cmap="gray")
plt.show()
```

```python
print(img.min(), img.mean(), img.max())
```

```python
img_neg = img
```

```python
plt.imshow(img_neg, cmap="gray")
```

```python
linha = img_neg.shape[0]
coluna = img_neg.shape[1]
```

```python
print(linha, coluna)
```

```python
for i in range(linha):
    for j in range(coluna):
        img_neg[i][j] = 255 - img_neg[i][j]
```

```python
plt.imshow(img_neg, cmap="gray")
```

```python
img_tr = img_neg
```

```python
img_tr[img_tr > 180] = 180
```

```python
img_tr[img_tr < 120] = 120
```

```python
plt.imshow(img_tr, cmap="gray")
```
