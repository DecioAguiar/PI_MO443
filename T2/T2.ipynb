```python
def str2bin(entrada):
    return ''.join([bin(ord(i))[2:].zfill(8) for i in entrada])
```

```python
def separar_camadas(img):
    R = img[:,:,0]
    G = img[:,:,1]
    B = img[:,:,2]
    
    return R,G,B
```

```python
def alterar_bit(key, bit, alt):
    x = list(bin(key))
    x = x[::-1]
    x[bit] = alt
    return ''.join(x[::-1])
```

```python
def recuperar_bit(key, bit):
    x = list(bin(key)[2:].zfill(8))
    x = x[::-1]
    return x[bit]
```

```python
def encode(img, msg, bit):
    bin_msg = str2bin(msg)
    
    dados = img.shape
    linhas = dados[0]
    colunas = dados[1]
    
    cont = 0
    for i in range(linhas):
        for j in range(colunas):
            for k in range(3):
                if(cont < len(bin_msg)):                                                
                    img[i,j,k] = int(alterar_bit(img[i,j,k], bit, bin_msg[cont]),2)                                        
                    cont += 1                    
    return img
```

```python
def bin2str(n):
    n = int('0b'+n,2)
    return chr(n)
```

```python
def txt_original(texto):
    return ''.join( [bin2str(texto[i:i+8]) for i in range(0,len(texto),8)])
```

```python
def decode(img, bit):
    
    dados = img.shape
    linhas = dados[0]
    colunas = dados[1]
    
    texto = ""
    texto_original = ""
    cont = 0
    for i in range(linhas):
        for j in range(colunas):
            for k in range(3):                                           
                texto = texto + recuperar_bit(img[i,j,k], bit)                                                                   
    return txt_original(texto)
```
