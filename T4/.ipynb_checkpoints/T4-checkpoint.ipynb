```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import math as mt
%matplotlib inline
import cv2
```

```python
img = rgb2gray(cv2.imread("baboon.pgm"))
```

```python
plt.imshow(img, cmap="gray")
plt.show()
```

```python
def vizinho_mais_proximo(img,x_,y_):
    return img[round(x_), round(y_)]
```

```python
def interpolacao_vmp(img, escala):
    nova_img = np.ones((escala*img.shape[0],escala*img.shape[1])) 
    linhas = nova_img.shape[0]
    colunas = nova_img.shape[1]
    for l in range(linhas-1):
        for c in range(colunas-1):
            nova_img[l,c] = vizinho_mais_proximo(img,(l/escala), (c/escala))
    return nova_img
```

```python
nova = interpolacao_vmp(img, 2)
```

```python
plt.imshow(nova, cmap="gray")
plt.show()
```
